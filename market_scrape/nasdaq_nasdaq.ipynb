```python
def scrape_nasdaq(symbol: str):
    HEADERS = {
        "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 \
                (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36",
        # "Accept-Language": "en-US, en;q=0.5",
    }
    URL = f"https://www.nasdaq.com/market-activity/stocks/{symbol.lower()}"

    # resp = requests.get(url=URL, headers=HEADERS, timeout=15)
    resp = requests.get(url=URL, timeout=10)
    soup = BeautifulSoup(resp.content, "html.parser")
    dom = etree.HTML(str(soup))

    market_cap = dom.xpath("/html/body/div[3]/div/main/div[2]/div[6]/div/div[1]/div/div[2]/table/tbody[2]/tr[1]/td[2]")[0].text
    # market_cap = dom.xpath('/html/body/div[4]/div/table[2]/tbody/tr[2]/td[2]/b')[0].text
    # market_cap = soup.select_one(
    #     "body > div.dialog-off-canvas-main-canvas > div > main > div.page__content > div:nth-child(6) > div > div.quote-detail__top-grid-data > div > div.summary-data__table-container.loaded > table > tbody:nth-child(3) > tr:nth-child(1) > td.summary-data__cell"
    # ).get_text()
    print(market_cap)
```

```python
symbols = ["AAPL", "TSLA", "MSFT"]
for symbol in symbols:
    scrape_nasdaq(symbol)
```
